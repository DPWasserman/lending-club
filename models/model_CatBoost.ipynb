## Set up data set

```python
import pandas as pd
import numpy as np

import sys
sys.path.append('../lending_club') # path the the directory
import config
from model_prep import divide_by_term, split_data
```

```python
accepted_loans = pd.read_parquet(config.DATAPATH / 'approved.parquet', engine='fastparquet')
accepted_36 = divide_by_term(accepted_loans,term_length=36)
```

```python
X_train, X_test, y_train, y_test = split_data(accepted_36.drop(['emp_title','addr_state'],axis=1), random_state=42)
```

```python
dummify_vars = config.VARS_TO_DUMMIFY.copy()
dummify_vars.remove('emp_title')
dummify_vars.remove('addr_state')
```

## CatBoost Modeling 

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
```

```python
%%time
params = {'depth': [5, 10, 50],
          'learning_rate' : [0.01, 0.1, 0.5, 1],
          'l2_leaf_reg': [1, 4, 9],
          'iterations': [100, 300, 600]}
cbc = CatBoostClassifier(cat_features=dummify_vars)
cbc_grid = GridSearchCV(cbc, params, scoring="accuracy", cv = 5)
cbc_grid.fit(X_train, y_train)
```

```python
best_cbc = cbc_grid.best_estimator_
best_params = cbc_grid.best_params_
```

```python
best_cbc.fit(X_train, y_train)
best_cbc.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
rms = mean_squared_error(y_test, best_cbc.predict(X_test), squared=False)
print(rms)
```

```python
from sklearn.metrics import mean_squared_error
cbc = CatBoostClassifier(depth=4, 
                        iterations=1000, 
                        l2_leaf_reg=1, 
                        learning_rate=0.03,
                        cat_features=list(config.VARS_TO_DUMMIFY.keys()))
cbr.fit(X_train, y_train,verbose=0, plot=True)
print(f'Training R2: {cbr.score(X_train, y_train)}')
print(f'Testing R2: {cbr.score(X_test, y_test)}')
print(f'RMSE: {mean_squared_error(y_test, cbr.predict(X_test), squared=False)}')
```
